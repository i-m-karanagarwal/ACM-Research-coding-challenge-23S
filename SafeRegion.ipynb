```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("6 class csv.csv")
```

```python
df.head()
```

```python
def calc_inner_outer_boundary(T_eff, L_star):
    inner_bou = math.sqrt(L_star / 1.1)
    outer_bou = math.sqrt(L_star / 0.53)
    
    return inner_bou, outer_bou
```

```python
inner_boundaries = []
outer_boundaries = []

for i in df.index:
    inner_boundary, outer_boundary = calc_inner_outer_boundary(df["T_eff"][i], df["L_star"][i])
    inner_boundaries.append(inner_boundary)
    outer_boundaries.append(outer_boundary)
    
```

```python
df['inner_boundary'] = inner_boundaries
df["outer_boundary"] = outer_boundaries
```
